## Slide 3: Feature Scaling

### What is Feature Scaling?

Feature scaling means adjusting feature values to a similar range using methods like:
- **Standardization**
- **Normalization**

### Why do we use it?

- It prevents large-value features from dominating smaller-value features.
- It improves performance for distance-based and gradient-based algorithms (e.g., KNN, SVM, Gradient Descent).
- It helps training converge faster.

### Example (Before vs After Scaling)

| Student | Height (cm) - Before | Weight (kg) - Before | Height (Scaled) | Weight (Scaled) |
|:--|--:|--:|--:|--:|
| A | 150 | 50 | 0.25 | 0.22 |
| B | 170 | 70 | 0.58 | 0.67 |
| C | 185 | 85 | 0.83 | 1.00 |

- Before scaling, height and weight have different numeric ranges.
- After scaling, both features are on a comparable scale and contribute more fairly.

## Slide 4: Feature Scaling Methods

### 1) Normalization (Min-Max Scaling)

Normalization transforms each value to a range between **0 and 1**.

$$
x_{\text{norm}} = \frac{x - x_{\min}}{x_{\max} - x_{\min}}
$$

**When to use it:**
- Useful for models that depend on distances or absolute magnitudes.
- Common choice for algorithms such as **KNN** and **K-Means**.

---

### 2) Standardization (Z-score Scaling)

Standardization centers data around **0** with a standard deviation of **1**.

$$
z = \frac{x - \mu}{\sigma}
$$

Where:
- $\mu$ = mean of the feature
- $\sigma$ = standard deviation of the feature

**When to use it:**
- Preferred when features are close to a Gaussian distribution.
- Works well with models like **Logistic Regression**, **SVM**, and **Linear Regression**.

## Slide 5: Choosing the Right Scaling Method

### Normalization

Use **Normalization** when feature values have very different ranges and you want all values mapped to **[0, 1]**.

- Typical method: **Min-Max Scaling**.
- Good for distance-based models where magnitude differences strongly affect results.

---

### Standardization

Use **Standardization** when features are expected to follow (or be close to) a **Gaussian distribution**, or when the dataset contains **outliers**.

- It keeps the data centered around 0 and scales variability to 1.
- Often preferred for linear models and optimization-based algorithms.

## Slide 8: Categorical Data

### What is Categorical Data?

Categorical data describes values that belong to **named groups** rather than continuous numbers.

### Category Types (Table)

| Type | Description | Example |
|:--|:--|:--|
| Ordinal | Categories have a meaningful order. | Education Level: High School < Bachelor's < Master's < PhD |
| Nominal | Categories have no natural order. | Payment Method: Cash, Card, Bank Transfer |

### Why Do We Encode Categorical Data?

Most machine learning algorithms work with **numerical inputs only**, so categorical values must be converted into numbers.

### Example 1: Label Encoding (Separate Table)

| Raw Category | Encoded Value |
|:--|:--:|
| Cash | 0 |
| Card | 1 |
| Bank Transfer | 2 |

### Example 2: One-Hot Encoding (Separate Table)

| Raw Category | Cash | Card | Bank Transfer |
|:--|:--:|:--:|:--:|
| Cash | 1 | 0 | 0 |
| Card | 0 | 1 | 0 |
| Bank Transfer | 0 | 0 | 1 |

## Slide 9: How to Deal with Categorical Data (Ordinal Encoding)

### What is Ordinal Encoding?

Ordinal Encoding converts ordered categories into integers that preserve their ranking.

### When to Use It

Use Ordinal Encoding only when categories have a **true order**.

### Steps

1. Identify the correct category order.
2. Assign an integer to each level based on that order.
3. Replace the original text values with encoded numbers.

### Example

| Satisfaction Level | Encoded Value |
|:--|:--:|
| Low | 1 |
| Medium | 2 |
| High | 3 |

This method keeps the order information, which is important for many machine learning models.

## Slide 10: How to Deal with Categorical Data (One-Hot Encoding)

### What is One-Hot Encoding?

One-Hot Encoding converts each category into a separate binary column (0 or 1).

### When to Use It

Use One-Hot Encoding for **nominal categories** (no natural order).

### Steps

1. List all unique categories.
2. Create one column for each category.
3. Put **1** in the matching category column and **0** in all others.

### Example

| Payment Method | Cash | Card | Bank Transfer |
|:--|:--:|:--:|:--:|
| Cash | 1 | 0 | 0 |
| Card | 0 | 1 | 0 |
| Bank Transfer | 0 | 0 | 1 |

This method avoids creating a false ranking between categories.

## Slide 11: How to Deal with Categorical Data (Dummy Encoding)

### What is Dummy Encoding?

Dummy Encoding is similar to One-Hot Encoding, but it removes one category column to avoid redundancy.

### Why Use It?

It helps prevent the **dummy variable trap** (perfect multicollinearity), especially in linear models.

### Steps

1. Start with One-Hot encoded columns.
2. Drop one category as a reference (baseline).
3. Keep the remaining binary columns.

### Example

Original categories: **Payment Method = [Cash, Card, Bank Transfer]**

Reference category (dropped): **Cash**

| Payment Method | Card | Bank Transfer |
|:--|:--:|:--:|
| Cash | 0 | 0 |
| Card | 1 | 0 |
| Bank Transfer | 0 | 1 |

In this setup, `Cash` is represented when all remaining columns are 0.

## Slide 12: How to Deal with Categorical Data (Other Encoding Types)

Besides Ordinal, One-Hot, and Dummy Encoding, there are other useful methods:

| Encoding Type | Idea | Best Use Case |
|:--|:--|:--|
| Frequency Encoding | Replace each category with how often it appears. | Large datasets with many categories |
| Count Encoding | Similar to frequency, but uses raw counts directly. | Tree-based models with high-cardinality features |
| Target Encoding | Replace category with target mean for that category. | Supervised tasks (use with care to avoid leakage) |
| Binary Encoding | Convert category index to binary digits across columns. | High-cardinality data with fewer columns than one-hot |
| Hash Encoding | Use a hash function to map categories into fixed columns. | Very large and dynamic category sets |

### Example for Each Type

Assume a feature `City` with values:

`[Riyadh, Jeddah, Riyadh, Dammam, Jeddah, Riyadh]`

#### 1) Frequency Encoding

| City | Frequency |
|:--|:--:|
| Riyadh | 0.50 |
| Jeddah | 0.33 |
| Dammam | 0.17 |

#### 2) Count Encoding

| City | Count |
|:--|:--:|
| Riyadh | 3 |
| Jeddah | 2 |
| Dammam | 1 |

#### 3) Target Encoding

| City | Mean Target |
|:--|:--:|
| Riyadh | 0.80 |
| Jeddah | 0.40 |
| Dammam | 0.20 |

#### 4) Binary Encoding

| City | Category ID | Binary Code |
|:--|:--:|:--:|
| Riyadh | 1 | 01 |
| Jeddah | 2 | 10 |
| Dammam | 3 | 11 |

#### 5) Hash Encoding (4 Buckets)

| City | Hash Bucket |
|:--|:--:|

| Dammam | 3 |

### Note

Choose the encoding method based on:
- whether categories are ordered,
- number of unique values (cardinality),
- model type,
- and risk of data leakage.

## Slide 13: Handling Missing Data

### What is Missing Data?

Missing data refers to values that are absent or incomplete in one or more features of a dataset.

### Types of Missing Data

| Type | Meaning |
|:--|:--|
| MCAR (Missing Completely at Random) | Missingness happens randomly and is unrelated to any variable in the dataset. |
| MAR (Missing at Random) | Missingness depends on other observed variables, but not on the missing value itself. |
| NMAR (Not Missing at Random) | Missingness depends on the missing value itself or unobserved factors. |

### Why Is It a Challenge?

- It can introduce bias into model training.
- It may reduce prediction accuracy.
- It can cause errors in analysis if not handled properly.

## Slide 14: Handling Missing Data - Why Do We Have Missing Data?

Missing values can appear for many practical reasons during data collection and processing.

| Cause | Description | Example |
|:--|:--|:--|
| Human Error | Data is skipped, entered incorrectly, or forgotten. | A user leaves the "Age" field empty in a form. |
| Device/Sensor Failure | Measurement tools fail or disconnect. | A medical sensor stops recording heart rate temporarily. |
| Data Integration Issues | Missing fields appear when combining data from multiple sources. | One database has `salary`, another does not. |
| Privacy or Refusal | Participants choose not to share sensitive information. | A customer does not provide income details. |
| System/Transmission Errors | Data is lost during storage, transfer, or export. | Network interruption causes missing rows in logs. |
| Conditional Missingness | Some fields are only relevant for specific groups. | `Pregnancy_Status` is empty for male patients. |

### Key Point

Understanding *why* data is missing helps us choose the right treatment method (deletion, imputation, or advanced modeling).

## Slide 15: Handling Missing Data - MCAR

### 1) Missing Completely at Random (MCAR)

MCAR means that missing values occur randomly and are **independent** of:
- observed variables, and
- unobserved variables.

So, the missingness has **no systematic pattern**.

### Important Reality Check

MCAR is often an **unrealistic assumption** in real-world datasets, because missing values are usually influenced by other factors.

### Common Influencing Factors

- **Human behavior:** people may skip sensitive questions.
- **Survey/process errors:** data collection mistakes can create missing entries.
- **External events:** unexpected conditions can affect responses.
- **Selective missingness:** some groups are more likely to leave fields blank.

### Why Context Is Essential

Before assuming MCAR, we should:
- understand the full data collection workflow,
- consult domain experts,
- and collaborate across data science and data engineering teams.

### Practical Approach

If MCAR appears reasonable after analysis and expert review, simple imputation methods (e.g., mean/median/mode) can be applied.

### Data Example (Illustrative)

| Student ID | Age | Score |
|:--:|:--:|:--:|
| 101 | 20 | 85 |
| 102 | *Missing* | 88 |
| 103 | 21 | *Missing* |
| 104 | 23 | 91 |
| 105 | *Missing* | 76 |
| 106 | 22 | *Missing* |

In this example, missing values appear across different rows and columns with no clear pattern, which is consistent with MCAR.

## Slide 16: Handling Missing Data - MAR

### 2) Missing at Random (MAR)

Missing at Random (MAR) means the probability of missing values is related to **observed features** in the dataset, but not to the missing feature itself.

### Practical Observation

MAR is more common in real-world datasets than MCAR.
Because missingness can be explained by available variables, we can often estimate it and apply reliable imputation.

### How to Identify MAR

Check whether the chance of missingness changes across other observed variables.
If it does, the pattern may be MAR.

### Example (Table)

| Student | Grade (Observed) | Study Hours |
|:--:|:--:|:--:|
| 1 | 95 | *Missing* |
| 2 | 92 | *Missing* |
| 3 | 88 | 4 |
| 4 | 75 | 6 |
| 5 | 70 | 7 |

In this example, students with higher grades are more likely to have missing `Study Hours`. Since missingness is related to `Grade` (an observed feature), this is consistent with **MAR**.

### Imputation Techniques for MAR

| Method | Idea | Why It Works for MAR |
|:--|:--|:--|
| kNN Imputation | Fills missing values using the nearest similar records. | Uses observed features to find similar patterns. |
| MissForest | Uses random-forest models iteratively to predict missing values. | Captures non-linear relationships from observed data effectively. |

### Example: kNN Imputation (k = 2)

Suppose we have this data:

| Student | Grade | Attendance (%) | Study Hours |
|:--:|:--:|:--:|:--:|
| A | 90 | 95 | 8 |
| B | 88 | 92 | 7 |
| C | 91 | 96 | *Missing* |
| D | 70 | 75 | 3 |

For student **C**, the two nearest students based on observed features (`Grade`, `Attendance`) are **A** and **B**.

kNN imputes the missing value using their average:

$$
\text{Study Hours}_C = \frac{8 + 7}{2} = 7.5
$$

So the imputed value for student **C** is **7.5** hours.

### Example: MissForest

Suppose we have the same dataset:

| Student | Grade | Attendance (%) | Study Hours |
|:--:|:--:|:--:|:--:|
| A | 90 | 95 | 8 |
| B | 88 | 92 | 7 |
| C | 91 | 96 | *Missing* |
| D | 70 | 75 | 3 |

MissForest trains a **Random Forest** model using rows with known `Study Hours` to predict the missing value for student **C** based on `Grade` and `Attendance`.

Predicted value (example):

$$
\text{Study Hours}_C \approx 7.6
$$

So, MissForest imputes student **C** with approximately **7.6** hours.

> Note: MissForest can capture complex non-linear relationships better than simple averaging methods.

## Slide 17: Handling Missing Data - MNAR

### 3) Missing Not at Random (MNAR)

MNAR is usually the most complex missing-data type.

In MNAR, missingness is related to:
- the missing value itself, or
- unobserved variables that were not collected.

Unlike MCAR (no pattern) and MAR (depends on observed features), MNAR has a systematic pattern tied to the missing variable.

### Why It Is Challenging

Because missingness depends on hidden or missing information, standard imputation can be unreliable without:
- additional data,
- strong domain knowledge,
- or careful modeling assumptions.

### Example (Table)

| Participant | Reported Stress Level |
|:--:|:--:|
| 1 | 2 |
| 2 | *Missing* |
| 3 | 3 |
| 4 | *Missing* |
| 5 | 1 |

Assume participants with **higher stress** are more likely to skip this question. Then missingness depends on the true (missing) stress value itself, which is MNAR.

### Practical Tip: Preserve Missingness Pattern

Create a binary indicator to show whether a value was missing and later imputed.

| Participant | Stress (Imputed) | Stress_Was_Missing |
|:--:|:--:|:--:|
| 1 | 2.0 | 0 |
| 2 | 4.1 | 1 |
| 3 | 3.0 | 0 |
| 4 | 4.3 | 1 |
| 5 | 1.0 | 0 |

This helps models learn useful signals from the missingness pattern itself.

## Slide 18: Data Preprocessing Template

### Standard Workflow

Use the following sequence to prepare data before training a machine learning model:

1. **Load the dataset**.
2. **Detect and handle missing values**.
3. **Encode categorical features** into numerical form.
4. **Split the dataset** into training and test sets.
5. **Scale numerical features** when needed.

### Pipeline Overview

**Raw Data -> Missing Data Treatment -> Categorical Encoding -> Train/Test Split -> Feature Scaling -> ML Model**

### Why This Order?

This order helps prevent data leakage and ensures that the model learns from clean, consistent, and properly transformed inputs.

## Slide 19: Regression - Simple Linear Regression

### What is Simple Linear Regression?

Simple Linear Regression models the relationship between:
- one independent variable `x` (input), and
- one dependent variable `y` (output),

using a straight line.

### General Equation

$$
y = b_0 + b_1 x
$$

### Meaning of Each Symbol

| Symbol | Meaning |
|:--|:--|
| $y$ | Predicted output (dependent variable). |
| $x$ | Input feature (independent variable). |
| $b_0$ | Intercept: predicted value of $y$ when $x=0$. |
| $b_1$ | Slope: expected change in $y$ when $x$ increases by 1 unit. |

### Practical Example

Suppose:
- `x` = Hours of Study
- `y` = Exam Score

Fitted equation:

$$
y = 40 + 5x
$$

Interpretation:
- **Slope ($b_1=5$):** each extra study hour increases the predicted score by 5 points.
- **Intercept ($b_0=40$):** if study hours are 0, predicted score is 40.

### Plot Interpretation

- Each point represents one student observation.
- The regression line is the best-fit line that minimizes prediction errors (distance between points and the line).

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data: x = study hours, y = exam score
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([45, 52, 55, 61, 66, 71, 74])

# Regression equation: y = 40 + 5x
b0 = 40
b1 = 5
y_pred = b0 + b1 * x

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='royalblue', s=70, label='Observed Data')
plt.plot(x, y_pred, color='gray', linewidth=2.5, label='Regression Line: y = 40 + 5x')

plt.title('Simple Linear Regression', fontsize=13)
plt.xlabel('Study Hours (x)')
plt.ylabel('Exam Score (y)')
plt.grid(alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```

## Slide 20: Ordinary Least Squares (OLS)

### What is OLS?

Ordinary Least Squares is the most common method used to fit a linear regression line.
It finds the line that makes prediction errors as small as possible.

### Main Idea

For each data point:
- Actual value: $y_i$
- Predicted value: $\hat{y}_i$
- Error (residual): $e_i = y_i - \hat{y}_i$

OLS chooses model parameters that minimize the **sum of squared residuals**.

### Objective Function

$$
\min \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

### Why Squared Errors?

- Makes all errors positive.
- Penalizes large errors more strongly.
- Gives a single clear objective for optimization.

### In Simple Linear Regression

If $\hat{y} = b_0 + b_1x$, OLS finds the best values of $b_0$ and $b_1$ that minimize total squared error.

## Slide 21: Multiple Linear Regression

### Definition

Multiple Linear Regression describes the relationship between:
- one dependent variable $y$, and
- two or more independent variables $(x_1, x_2, ..., x_n)$.

It models this relationship using a linear equation.

### Model Equation

$$
y = b_0 + b_1x_1 + b_2x_2 + \cdots + b_nx_n
$$

### Purpose

The main goal is to predict the target variable $y$ using multiple input factors.

### Why It Is Useful

Compared with simple linear regression, this model captures more realistic scenarios where outcomes are affected by several variables at the same time.

### Example

Suppose we want to predict house price:

- $y$ = House Price (in $1000)
- $x_1$ = House Size (m^2)
- $x_2$ = Number of Bedrooms

A fitted model could be:

$$
y = 50 + 2.5x_1 + 15x_2
$$

Interpretation:
- For each additional 1 m^2, price increases by **2.5** (holding other variables constant).
- For each additional bedroom, price increases by **15** (holding other variables constant).

### Table for Clarification

| House | Size ($x_1$) m^2 | Bedrooms ($x_2$) | Predicted Price ($y$ in $1000) |
|:--:|:--:|:--:|:--:|
| A | 100 | 2 | $50 + 2.5(100) + 15(2) = 330$ |
| B | 120 | 3 | $50 + 2.5(120) + 15(3) = 395$ |
| C | 80  | 1 | $50 + 2.5(80) + 15(1) = 265$ |

## Slide 22: R-Squared ($R^2$)

### Definition

$R^2$ is a goodness-of-fit metric in regression.
It tells us how much of the variation in the target variable $y$ is explained by the model inputs.

### Equation

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

### Formulas for $SS_{res}$ and $SS_{tot}$

$$
SS_{res} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

$$
SS_{tot} = \sum_{i=1}^{n} (y_i - \bar{y})^2
$$

### What Do These Terms Mean?

| Term | Description |
|:--|:--|
| $SS_{res}$ | Residual Sum of Squares: total squared distance between actual values ($y_i$) and predicted values ($\hat{y}_i$). |
| $SS_{tot}$ | Total Sum of Squares: total squared distance between actual values ($y_i$) and the mean of $y$ ($\bar{y}$). |

### How to Read $R^2$

- A larger $R^2$ means a better explanatory power.
- $R^2 = 1$ means predictions perfectly match the data.
- $R^2 = 0$ means the model performs like a simple mean-based prediction.

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2.2, 2.8, 3.6, 4.5, 5.1, 5.9])

# Fitted regression line (example)
y_hat = 1.4 + 0.75 * x
y_bar = np.mean(y)

# Sums of squares
ss_res = np.sum((y - y_hat) ** 2)
ss_tot = np.sum((y - y_bar) ** 2)
r2 = 1 - (ss_res / ss_tot)

# -----------------------------
# Figure 1: SS_res only
# -----------------------------
plt.figure(figsize=(8.5, 5))
plt.scatter(x, y, color='royalblue', s=65, label='Actual values ($y_i$)', zorder=3)
plt.plot(x, y_hat, color='dimgray', linewidth=2.2, label='Predicted line ($\\hat{y}_i$)', zorder=2)

for xi, yi, yhi in zip(x, y, y_hat):
    plt.plot([xi, xi], [yi, yhi], color='crimson', linewidth=2, alpha=0.9, linestyle='--')

plt.text(
    0.02,
    0.98,
    r"$SS_{res}=\sum (y_i-\hat{y}_i)^2$" + f"\nSS_res = {ss_res:.3f}",
    transform=plt.gca().transAxes,
    va='top',
    ha='left',
    fontsize=10,
    bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.9, edgecolor='gray')
)

plt.plot([], [], color='crimson', linewidth=2, label='Residual distance ($y_i-\\hat{y}_i$)')
plt.title('Residual Sum of Squares ($SS_{res}$)', fontsize=13)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.22)
plt.legend(loc='lower right', fontsize=9)
plt.tight_layout()
plt.show()

# -----------------------------
# Figure 2: SS_tot only
# -----------------------------
plt.figure(figsize=(8.5, 5))
plt.scatter(x, y, color='royalblue', s=65, label='Actual values ($y_i$)', zorder=3)
plt.axhline(y=y_bar, color='green', linestyle='--', linewidth=1.8, label='Mean of $y$ ($\\bar{y}$)')

for xi, yi in zip(x, y):
    plt.plot([xi, xi], [yi, y_bar], color='darkorange', linewidth=2, alpha=0.85)

plt.text(
    0.02,
    0.98,
    r"$SS_{tot}=\sum (y_i-\bar{y})^2$" + f"\nSS_tot = {ss_tot:.3f}\n$R^2$ = {r2:.3f}",
    transform=plt.gca().transAxes,
    va='top',
    ha='left',
    fontsize=10,
    bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.9, edgecolor='gray')
)

plt.plot([], [], color='darkorange', linewidth=2, label='Distance from mean ($y_i-\\bar{y}$)')
plt.title('Total Sum of Squares ($SS_{tot}$)', fontsize=13)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.22)
plt.legend(loc='lower right', fontsize=9)
plt.tight_layout()
plt.show()
```
