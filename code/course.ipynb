## Slide 3: Feature Scaling

### What is Feature Scaling?

Feature scaling means adjusting feature values to a similar range using methods like:
- **Standardization**
- **Normalization**

### Why do we use it?

- It prevents large-value features from dominating smaller-value features.
- It improves performance for distance-based and gradient-based algorithms (e.g., KNN, SVM, Gradient Descent).
- It helps training converge faster.

### Example (Before vs After Scaling)

| Student | Height (cm) - Before | Weight (kg) - Before | Height (Scaled) | Weight (Scaled) |
|:--|--:|--:|--:|--:|
| A | 150 | 50 | 0.25 | 0.22 |
| B | 170 | 70 | 0.58 | 0.67 |
| C | 185 | 85 | 0.83 | 1.00 |

- Before scaling, height and weight have different numeric ranges.
- After scaling, both features are on a comparable scale and contribute more fairly.

## Slide 4: Feature Scaling Methods

### 1) Normalization (Min-Max Scaling)

Normalization transforms each value to a range between **0 and 1**.

$$
x_{\text{norm}} = \frac{x - x_{\min}}{x_{\max} - x_{\min}}
$$

**When to use it:**
- Useful for models that depend on distances or absolute magnitudes.
- Common choice for algorithms such as **KNN** and **K-Means**.

---

### 2) Standardization (Z-score Scaling)

Standardization centers data around **0** with a standard deviation of **1**.

$$
z = \frac{x - \mu}{\sigma}
$$

Where:
- $\mu$ = mean of the feature
- $\sigma$ = standard deviation of the feature

**When to use it:**
- Preferred when features are close to a Gaussian distribution.
- Works well with models like **Logistic Regression**, **SVM**, and **Linear Regression**.

## Slide 5: Choosing the Right Scaling Method

### Normalization

Use **Normalization** when feature values have very different ranges and you want all values mapped to **[0, 1]**.

- Typical method: **Min-Max Scaling**.
- Good for distance-based models where magnitude differences strongly affect results.

---

### Standardization

Use **Standardization** when features are expected to follow (or be close to) a **Gaussian distribution**, or when the dataset contains **outliers**.

- It keeps the data centered around 0 and scales variability to 1.
- Often preferred for linear models and optimization-based algorithms.

## Slide 8: Categorical Data

### What is Categorical Data?

Categorical data describes values that belong to **named groups** rather than continuous numbers.

### Category Types (Table)

### Why Do We Encode Categorical Data?

Most machine learning algorithms work with **numerical inputs only**, so categorical values must be converted into numbers.

## Slide 9: How to Deal with Categorical Data (Ordinal Encoding)

### What is Ordinal Encoding?

Ordinal Encoding converts ordered categories into integers that preserve their ranking.

### When to Use It

Use Ordinal Encoding only when categories have a **true order**.

### Steps

1. Identify the correct category order.
2. Assign an integer to each level based on that order.
3. Replace the original text values with encoded numbers.

### Example

| Satisfaction Level | Encoded Value |
|:--|:--:|
| Low | 1 |
| Medium | 2 |
| High | 3 |

This method keeps the order information, which is important for many machine learning models.

## Slide 10: How to Deal with Categorical Data (One-Hot Encoding)

### What is One-Hot Encoding?

One-Hot Encoding converts each category into a separate binary column (0 or 1).

### When to Use It

Use One-Hot Encoding for **nominal categories** (no natural order).

### Steps

1. List all unique categories.
2. Create one column for each category.
3. Put **1** in the matching category column and **0** in all others.

### Example

| Payment Method | Cash | Card | Bank Transfer |
|:--|:--:|:--:|:--:|
| Cash | 1 | 0 | 0 |
| Card | 0 | 1 | 0 |
| Bank Transfer | 0 | 0 | 1 |

This method avoids creating a false ranking between categories.

## Slide 11: How to Deal with Categorical Data (Dummy Encoding)

### What is Dummy Encoding?

Dummy Encoding is similar to One-Hot Encoding, but it removes one category column to avoid redundancy.

### Why Use It?

It helps prevent the **dummy variable trap** (perfect multicollinearity), especially in linear models.

### Steps

1. Start with One-Hot encoded columns.
2. Drop one category as a reference (baseline).
3. Keep the remaining binary columns.

### Example

Original categories: **Payment Method = [Cash, Card, Bank Transfer]**

Reference category (dropped): **Cash**

| Payment Method | Card | Bank Transfer |
|:--|:--:|:--:|
| Cash | 0 | 0 |
| Card | 1 | 0 |
| Bank Transfer | 0 | 1 |

In this setup, `Cash` is represented when all remaining columns are 0.

## Slide 12: How to Deal with Categorical Data (Other Encoding Types)

Besides Ordinal, One-Hot, and Dummy Encoding, there are other useful methods:

| Encoding Type | Idea | Best Use Case |
|:--|:--|:--|
| Frequency Encoding | Replace each category with how often it appears. | Large datasets with many categories |
| Count Encoding | Similar to frequency, but uses raw counts directly. | Tree-based models with high-cardinality features |
| Target Encoding | Replace category with target mean for that category. | Supervised tasks (use with care to avoid leakage) |
| Binary Encoding | Convert category index to binary digits across columns. | High-cardinality data with fewer columns than one-hot |
| Hash Encoding | Use a hash function to map categories into fixed columns. | Very large and dynamic category sets |

## Slide 13: Handling Missing Data

### What is Missing Data?

Missing data refers to values that are absent or incomplete in one or more features of a dataset.

### Types of Missing Data

| Type | Meaning |
|:--|:--|
| MCAR (Missing Completely at Random) | Missingness happens randomly and is unrelated to any variable in the dataset. |
| MAR (Missing at Random) | Missingness depends on other observed variables, but not on the missing value itself. |
| NMAR (Not Missing at Random) | Missingness depends on the missing value itself or unobserved factors. |

### Why Is It a Challenge?

- It can introduce bias into model training.
- It may reduce prediction accuracy.
- It can cause errors in analysis if not handled properly.

## Slide 14: Handling Missing Data - Why Do We Have Missing Data?

Missing values can appear for many practical reasons during data collection and processing.

| Cause | Description | Example |
|:--|:--|:--|
| Human Error | Data is skipped, entered incorrectly, or forgotten. | A user leaves the "Age" field empty in a form. |
| Device/Sensor Failure | Measurement tools fail or disconnect. | A medical sensor stops recording heart rate temporarily. |
| Data Integration Issues | Missing fields appear when combining data from multiple sources. | One database has `salary`, another does not. |
| Privacy or Refusal | Participants choose not to share sensitive information. | A customer does not provide income details. |
| System/Transmission Errors | Data is lost during storage, transfer, or export. | Network interruption causes missing rows in logs. |
| Conditional Missingness | Some fields are only relevant for specific groups. | `Pregnancy_Status` is empty for male patients. |

### Key Point

Understanding *why* data is missing helps us choose the right treatment method (deletion, imputation, or advanced modeling).

## Slide 15: Handling Missing Data - MCAR

### 1) Missing Completely at Random (MCAR)

MCAR means that missing values occur randomly and are **independent** of:
- observed variables, and
- unobserved variables.

So, the missingness has **no systematic pattern**.

### Important Reality Check

MCAR is often an **unrealistic assumption** in real-world datasets, because missing values are usually influenced by other factors.

### Common Influencing Factors

- **Human behavior:** people may skip sensitive questions.
- **Survey/process errors:** data collection mistakes can create missing entries.
- **External events:** unexpected conditions can affect responses.
- **Selective missingness:** some groups are more likely to leave fields blank.

### Why Context Is Essential

Before assuming MCAR, we should:
- understand the full data collection workflow,
- consult domain experts,
- and collaborate across data science and data engineering teams.

### Practical Approach

If MCAR appears reasonable after analysis and expert review, simple imputation methods (e.g., mean/median/mode) can be applied.

### Data Example (Illustrative)

| Student ID | Age | Score |
|:--:|:--:|:--:|
| 101 | 20 | 85 |
| 102 | *Missing* | 88 |
| 103 | 21 | *Missing* |
| 104 | 23 | 91 |
| 105 | *Missing* | 76 |
| 106 | 22 | *Missing* |

In this example, missing values appear across different rows and columns with no clear pattern, which is consistent with MCAR.

## Slide 16: Handling Missing Data - MAR

### 2) Missing at Random (MAR)

Missing at Random (MAR) means the probability of missing values is related to **observed features** in the dataset, but not to the missing feature itself.

### Practical Observation

MAR is more common in real-world datasets than MCAR.
Because missingness can be explained by available variables, we can often estimate it and apply reliable imputation.

### How to Identify MAR

Check whether the chance of missingness changes across other observed variables.
If it does, the pattern may be MAR.

### Example (Table)

| Student | Grade (Observed) | Study Hours |
|:--:|:--:|:--:|
| 1 | 95 | *Missing* |
| 2 | 92 | *Missing* |
| 3 | 88 | 4 |
| 4 | 75 | 6 |
| 5 | 70 | 7 |

In this example, students with higher grades are more likely to have missing `Study Hours`. Since missingness is related to `Grade` (an observed feature), this is consistent with **MAR**.

### Imputation Techniques for MAR

| Method | Idea | Why It Works for MAR |
|:--|:--|:--|
| kNN Imputation | Fills missing values using the nearest similar records. | Uses observed features to find similar patterns. |
| MissForest | Uses random-forest models iteratively to predict missing values. | Captures non-linear relationships from observed data effectively. |

### Example: kNN Imputation (k = 2)

Suppose we have this data:

| Student | Grade | Attendance (%) | Study Hours |
|:--:|:--:|:--:|:--:|
| A | 90 | 95 | 8 |
| B | 88 | 92 | 7 |
| C | 91 | 96 | *Missing* |
| D | 70 | 75 | 3 |

For student **C**, the two nearest students based on observed features (`Grade`, `Attendance`) are **A** and **B**.

kNN imputes the missing value using their average:

$$
\text{Study Hours}_C = \frac{8 + 7}{2} = 7.5
$$

So the imputed value for student **C** is **7.5** hours.

### Example: MissForest

Suppose we have the same dataset:

| Student | Grade | Attendance (%) | Study Hours |
|:--:|:--:|:--:|:--:|
| A | 90 | 95 | 8 |
| B | 88 | 92 | 7 |
| C | 91 | 96 | *Missing* |
| D | 70 | 75 | 3 |

MissForest trains a **Random Forest** model using rows with known `Study Hours` to predict the missing value for student **C** based on `Grade` and `Attendance`.

Predicted value (example):

$$
\text{Study Hours}_C \approx 7.6
$$

So, MissForest imputes student **C** with approximately **7.6** hours.

> Note: MissForest can capture complex non-linear relationships better than simple averaging methods.

## Slide 17: Handling Missing Data - MNAR

### 3) Missing Not at Random (MNAR)

MNAR is usually the most complex missing-data type.

In MNAR, missingness is related to:
- the missing value itself, or
- unobserved variables that were not collected.

Unlike MCAR (no pattern) and MAR (depends on observed features), MNAR has a systematic pattern tied to the missing variable.

### Why It Is Challenging

Because missingness depends on hidden or missing information, standard imputation can be unreliable without:
- additional data,
- strong domain knowledge,
- or careful modeling assumptions.

### Example (Table)

| Participant | Reported Stress Level |
|:--:|:--:|
| 1 | 2 |
| 2 | *Missing* |
| 3 | 3 |
| 4 | *Missing* |
| 5 | 1 |

Assume participants with **higher stress** are more likely to skip this question. Then missingness depends on the true (missing) stress value itself, which is MNAR.

### Practical Tip: Preserve Missingness Pattern

Create a binary indicator to show whether a value was missing and later imputed.

| Participant | Stress (Imputed) | Stress_Was_Missing |
|:--:|:--:|:--:|
| 1 | 2.0 | 0 |
| 2 | 4.1 | 1 |
| 3 | 3.0 | 0 |
| 4 | 4.3 | 1 |
| 5 | 1.0 | 0 |

This helps models learn useful signals from the missingness pattern itself.

## Slide 18: Data Preprocessing Template

### Standard Workflow

Use the following sequence to prepare data before training a machine learning model:

1. **Load the dataset**.
2. **Detect and handle missing values**.
3. **Encode categorical features** into numerical form.
4. **Split the dataset** into training and test sets.
5. **Scale numerical features** when needed.

### Pipeline Overview

**Raw Data -> Missing Data Treatment -> Categorical Encoding -> Train/Test Split -> Feature Scaling -> ML Model**

## Slide 19: Regression - Simple Linear Regression

### What is Simple Linear Regression?

Simple Linear Regression models the relationship between:
- one independent variable `x` (input), and
- one dependent variable `y` (output),

using a straight line.

### General Equation

$$
y = b_0 + b_1 x
$$

### Meaning of Each Symbol

| Symbol | Meaning |
|:--|:--|
| $y$ | Predicted output (dependent variable). |
| $x$ | Input feature (independent variable). |
| $b_0$ | Intercept: predicted value of $y$ when $x=0$. |
| $b_1$ | Slope: expected change in $y$ when $x$ increases by 1 unit. |

### Practical Example

Suppose:
- `x` = Hours of Study
- `y` = Exam Score

Fitted equation:

$$
y = 40 + 5x
$$

Interpretation:
- **Slope ($b_1=5$):** each extra study hour increases the predicted score by 5 points.
- **Intercept ($b_0=40$):** if study hours are 0, predicted score is 40.

### Plot Interpretation

- Each point represents one student observation.
- The regression line is the best-fit line that minimizes prediction errors (distance between points and the line).

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data: x = study hours, y = exam score
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([45, 52, 55, 61, 66, 71, 74])

# Regression equation: y = 40 + 5x
b0 = 40
b1 = 5
y_pred = b0 + b1 * x

plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='royalblue', s=70, label='Observed Data')
plt.plot(x, y_pred, color='gray', linewidth=2.5, label='Regression Line: y = 40 + 5x')

plt.title('Simple Linear Regression', fontsize=13)
plt.xlabel('Study Hours (x)')
plt.ylabel('Exam Score (y)')
plt.grid(alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```

## Slide 20: Ordinary Least Squares (OLS)

### What is OLS?

Ordinary Least Squares is the most common method used to fit a linear regression line.
It finds the line that makes prediction errors as small as possible.

### Main Idea

For each data point:
- Actual value: $y_i$
- Predicted value: $\hat{y}_i$
- Error (residual): $e_i = y_i - \hat{y}_i$

OLS chooses model parameters that minimize the **sum of squared residuals**.

### Objective Function

$$
\min \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

### Why Squared Errors?

- Makes all errors positive.
- Penalizes large errors more strongly.
- Gives a single clear objective for optimization.

### In Simple Linear Regression

If $\hat{y} = b_0 + b_1x$, OLS finds the best values of $b_0$ and $b_1$ that minimize total squared error.

## Slide 21: Multiple Linear Regression

### Definition

Multiple Linear Regression describes the relationship between:
- one dependent variable $y$, and
- two or more independent variables $(x_1, x_2, ..., x_n)$.

It models this relationship using a linear equation.

### Model Equation

$$
y = b_0 + b_1x_1 + b_2x_2 + \cdots + b_nx_n
$$

### Purpose

The main goal is to predict the target variable $y$ using multiple input factors.

### Why It Is Useful

Compared with simple linear regression, this model captures more realistic scenarios where outcomes are affected by several variables at the same time.

### Example

Suppose we want to predict house price:

- $y$ = House Price (in $1000)
- $x_1$ = House Size (m^2)
- $x_2$ = Number of Bedrooms

A fitted model could be:

$$
y = 50 + 2.5x_1 + 15x_2
$$

Interpretation:
- For each additional 1 m^2, price increases by **2.5** (holding other variables constant).
- For each additional bedroom, price increases by **15** (holding other variables constant).

### Table for Clarification

| House | Size ($x_1$) m^2 | Bedrooms ($x_2$) | Predicted Price ($y$ in $1000) |
|:--:|:--:|:--:|:--:|
| A | 100 | 2 | $50 + 2.5(100) + 15(2) = 330$ |
| B | 120 | 3 | $50 + 2.5(120) + 15(3) = 395$ |
| C | 80  | 1 | $50 + 2.5(80) + 15(1) = 265$ |

## Slide 22: R-Squared ($R^2$)

### Definition

$R^2$ is a goodness-of-fit metric in regression.
It tells us how much of the variation in the target variable $y$ is explained by the model inputs.

### Equation

$$
R^2 = 1 - \frac{SS_{res}}{SS_{tot}}
$$

### Formulas for $SS_{res}$ and $SS_{tot}$

$$
SS_{res} = \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

$$
SS_{tot} = \sum_{i=1}^{n} (y_i - \bar{y})^2
$$

### What Do These Terms Mean?

| Term | Description |
|:--|:--|
| $SS_{res}$ | Residual Sum of Squares: total squared distance between actual values ($y_i$) and predicted values ($\hat{y}_i$). |
| $SS_{tot}$ | Total Sum of Squares: total squared distance between actual values ($y_i$) and the mean of $y$ ($\bar{y}$). |

### How to Read $R^2$

- A larger $R^2$ means a better explanatory power.
- $R^2 = 1$ means predictions perfectly match the data.
- $R^2 = 0$ means the model performs like a simple mean-based prediction.

```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
x = np.array([1, 2, 3, 4, 5, 6])
y = np.array([2.2, 2.8, 3.6, 4.5, 5.1, 5.9])

# Fitted regression line (example)
y_hat = 1.4 + 0.75 * x
y_bar = np.mean(y)

# Sums of squares
ss_res = np.sum((y - y_hat) ** 2)
ss_tot = np.sum((y - y_bar) ** 2)
r2 = 1 - (ss_res / ss_tot)

# -----------------------------
# Figure 1: SS_res only
# -----------------------------
plt.figure(figsize=(8.5, 5))
plt.scatter(x, y, color='royalblue', s=65, label='Actual values ($y_i$)', zorder=3)
plt.plot(x, y_hat, color='dimgray', linewidth=2.2, label='Predicted line ($\\hat{y}_i$)', zorder=2)

for xi, yi, yhi in zip(x, y, y_hat):
    plt.plot([xi, xi], [yi, yhi], color='crimson', linewidth=2, alpha=0.9, linestyle='--')

plt.text(
    0.02,
    0.98,
    r"$SS_{res}=\sum (y_i-\hat{y}_i)^2$" + f"\nSS_res = {ss_res:.3f}",
    transform=plt.gca().transAxes,
    va='top',
    ha='left',
    fontsize=10,
    bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.9, edgecolor='gray')
)

plt.plot([], [], color='crimson', linewidth=2, label='Residual distance ($y_i-\\hat{y}_i$)')
plt.title('Residual Sum of Squares ($SS_{res}$)', fontsize=13)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.22)
plt.legend(loc='lower right', fontsize=9)
plt.tight_layout()
plt.show()

# -----------------------------
# Figure 2: SS_tot only
# -----------------------------
plt.figure(figsize=(8.5, 5))
plt.scatter(x, y, color='royalblue', s=65, label='Actual values ($y_i$)', zorder=3)
plt.axhline(y=y_bar, color='green', linestyle='--', linewidth=1.8, label='Mean of $y$ ($\\bar{y}$)')

for xi, yi in zip(x, y):
    plt.plot([xi, xi], [yi, y_bar], color='darkorange', linewidth=2, alpha=0.85)

plt.text(
    0.02,
    0.98,
    r"$SS_{tot}=\sum (y_i-\bar{y})^2$" + f"\nSS_tot = {ss_tot:.3f}\n$R^2$ = {r2:.3f}",
    transform=plt.gca().transAxes,
    va='top',
    ha='left',
    fontsize=10,
    bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.9, edgecolor='gray')
)

plt.plot([], [], color='darkorange', linewidth=2, label='Distance from mean ($y_i-\\bar{y}$)')
plt.title('Total Sum of Squares ($SS_{tot}$)', fontsize=13)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.22)
plt.legend(loc='lower right', fontsize=9)
plt.tight_layout()
plt.show()
```

## Slide 23: Adjusted R-Squared

### Why $R^2$ Alone Is Not Enough

In multiple regression, adding new predictors often makes $R^2$ increase (or stay the same), even when the new variables are not truly useful.

This happens because:
- $SS_{res}$ usually decreases when new predictors are added,
- while $SS_{tot}$ remains fixed for the same target variable.

So, a higher $R^2$ does not always mean a better model.

### Risk: Overfitting

Adding irrelevant predictors can make the model more complex without real improvement in generalization.

### Adjusted $R^2$ (Better Metric)

Adjusted $R^2$ adds a penalty for unnecessary predictors.

$$
\bar{R}^2 = 1 - (1 - R^2)\frac{n-1}{n-p-1}
$$

Where:
- $n$: number of observations
- $p$: number of predictors

### Quick Example

| Model | Predictors | $R^2$ | Adjusted $R^2$ | Interpretation |
|:--|:--|:--:|:--:|:--|
| A | $x_1, x_2$ | 0.82 | 0.81 | Strong and efficient model |
| B | $x_1, x_2, x_3, x_4$ | 0.83 | 0.79 | Slightly higher $R^2$, but worse after penalty |

Model B looks better by $R^2$, but Adjusted $R^2$ shows that extra variables are likely not helpful.

## Slide 24: Assumptions of Linear Regression (Overview)

Linear regression gives reliable results when key assumptions are approximately satisfied.

| Assumption | Meaning | If Violated |
|:--|:--|:--|
| Linearity | Relationship between $y$ and each $x$ is roughly straight-line. | Biased predictions |
| Homoscedasticity | Error variance is roughly constant across fitted values. | Unstable standard errors |
| Normality of Errors | Residuals are approximately normal. | Inference (tests/CIs) becomes less reliable |
| Independence | Observations/errors are independent. | Biased significance tests |
| No Severe Multicollinearity | Predictors are not highly redundant. | Coefficients become unstable |
| Limited Outlier Influence | Extreme points do not dominate the fit. | Distorted regression line |

## Slide 25: Assumptions with Practical Examples

| Assumption | Simple Example | Practical Check |
|:--|:--|:--|
| Linearity | Sales increase approximately linearly with ad spend. | Scatter plot / residual plot |
| Homoscedasticity | Prediction errors are similar for low and high sales values. | Residuals vs fitted values |
| Normal Errors | Most residuals are near zero, few at extremes. | Histogram / Q-Q plot of residuals |
| Independence | Customer records are independent (not repeated from same person). | Study design / Durbin-Watson for time data |
| Low Multicollinearity | `TV_ads` and `radio_ads` are not near-duplicates. | Correlation matrix / VIF |
| Outlier Control | One extreme house price should not define the whole model. | Cook's distance / leverage diagnostics |

### Note

If one assumption is violated, we can often fix it with transformations, robust methods, feature engineering, or a different model.

## Slide 26: Dummy Variables and Dummy Variable Trap

### Dummy Variables

A categorical variable is converted into binary columns (0/1).

Example: `State` has two categories: `New York`, `California`

| State | $D_1$ (New York) | $D_2$ (California) |
|:--|:--:|:--:|
| New York | 1 | 0 |
| California | 0 | 1 |

### Dummy Variable Trap

If all dummy columns are included with an intercept term, perfect multicollinearity appears:

$$
D_2 = 1 - D_1
$$

So one dummy can be fully predicted from the other.

### Why This Is a Problem

- Coefficients become unstable.
- Model interpretation becomes unreliable.
- Statistical tests may become misleading.

## Slide 27: Solving Dummy Variable Trap

### Solution

Drop one dummy column and keep it as the baseline category.

If we drop `California`:
- $D_1 = 1$ -> New York
- $D_1 = 0$ -> California (baseline)

### Regression Example

$$
\text{Profit} = b_0 + b_1x_1 + b_2x_2 + b_3x_3 + b_4D_1
$$

Where:
- $x_1, x_2, x_3$: numerical predictors (e.g., R&D, Admin, Marketing)
- $D_1$: state dummy

### Interpretation Example

| Case | $D_1$ | State Meaning |
|:--|:--:|:--|
| A | 1 | New York |
| B | 0 | California (reference) |

This keeps category information while avoiding multicollinearity.

## Slide 28: Building a Regression Model

### Goal

Model-building methods select the most relevant predictors to balance:
- simplicity,
- interpretability,
- and predictive accuracy.

| Method | Core Idea | Best Use Case |
|:--|:--|:--|
| All-in | Use all predictors directly. | Strong prior knowledge that all variables matter |
| Backward Elimination | Start full, remove least significant variables. | Many candidate predictors |
| Forward Selection | Start empty, add most significant variables. | Need a compact model from scratch |
| Bidirectional (Stepwise) | Add and remove dynamically. | Flexible search for balanced model |

## Slide 29: Model-Building Methods (Steps + Example)

### 1) Backward Elimination

1. Set significance level (e.g., $SL=0.05$)
2. Fit model with all predictors
3. Remove highest p-value if $p>SL$
4. Refit and repeat

### 2) Forward Selection

1. Start with no predictors
2. Add variable with lowest p-value below $SL$
3. Continue adding one-by-one
4. Stop when no variable qualifies

### 3) Bidirectional (Stepwise)

- Add significant variables and remove non-significant ones in each iteration.

### Mini Example

| Iteration | Candidate Predictors | Selected Action |
|:--|:--|:--|
| Start | $x_1,x_2,x_3,x_4$ | Full model |
| 1 | highest p-value: $x_4=0.42$ | Remove $x_4$ |
| 2 | highest p-value: $x_3=0.11$ | Remove $x_3$ |
| Final | $x_1,x_2$ with $p<0.05$ | Keep final model |

## Slide 30: Significance Level and p-value

| Concept | Meaning |
|:--|:--|
| Null Hypothesis ($H_0$) | No effect / no relationship |
| p-value | Probability of observing current result (or more extreme) if $H_0$ is true |
| Significance Level ($\alpha$) | Maximum tolerated Type I error probability |

### Decision Rule

- If $p < \alpha$: reject $H_0$ (statistically significant)
- If $p \geq \alpha$: fail to reject $H_0$

### Example

If $p=0.03$ and $\alpha=0.05$:

$$
0.03 < 0.05 \Rightarrow \text{Reject } H_0
$$

So the predictor is considered significant at the 5% level.

## Slide 31: Why Is $SL=0.05$ Common?

### Standard Choice

$SL=0.05$ is widely used because it balances sensitivity and false-positive risk.

### Relation to Confidence Level

$$
SL = 1 - \text{Confidence Level}
$$

| $SL$ | Confidence Level | Strictness |
|:--:|:--:|:--|
| 0.10 | 90% | Less strict |
| 0.05 | 95% | Standard |
| 0.01 | 99% | More strict |

### How to Choose in Practice

| Scenario | Suggested $SL$ |
|:--|:--:|
| High-stakes decisions (medicine, safety, finance) | 0.01 |
| General modeling and reporting | 0.05 |
| Exploratory analysis | 0.10 |

### Final Tip

Start with $SL=0.05$, then adjust based on domain risk, dataset size, and model purpose.

## Slide 33: Polynomial Regression

### Definition

Polynomial Regression is an extension of linear regression used to model **non-linear relationships** by adding polynomial terms of the input variable.

### Core Idea

- Linear regression: $y = b_0 + b_1x$
- Polynomial regression: $y = b_0 + b_1x + b_2x^2 + b_3x^3 + \dots$

Even when the curve is non-linear, the model is still called **linear in parameters** because it is linear with respect to the coefficients $(b_0, b_1, b_2, ...)$.

## Slide 34: Why Polynomial Regression?

### Why Not Always Linear Regression?

Simple linear regression fits only straight-line patterns, but many real-world relationships are curved.

### Typical Non-Linear Patterns

- Population growth over time
- Disease progression
- Sales/economic trends
- Rainfall vs crop yield relationships

### Example Insight

If data follows a U-shape or curved trend, a straight line will underfit, while polynomial terms can better capture the pattern.

## Slide 35: How Polynomial Regression Works

| Step | Description |
|:--|:--|
| 1 | Start with input feature $x$ |
| 2 | Create transformed terms: $x^2, x^3, ...$ |
| 3 | Fit linear regression on transformed features |
| 4 | Estimate coefficients $(b_0, b_1, b_2, ...)$ |
| 5 | Use the fitted equation to predict $y$ |

### Model Form

$$
y = b_0 + b_1x + b_2x^2 + \cdots + b_dx^d
$$

Where $d$ is the polynomial degree.

## Slide 36: Choosing the Polynomial Degree

The degree controls model complexity.

| Degree Choice | Effect |
|:--|:--|
| Too low | Underfitting (model too simple) |
| Too high | Overfitting (fits noise, weak generalization) |
| Balanced degree | Better bias-variance trade-off |

### Practical Selection Methods

- Cross-validation
- Adjusted $R^2$
- Validation/test error comparison

## Slide 37: When to Use Polynomial Regression

### Good Use Cases

- Data shows clear curved trends
- You need more flexibility than a straight line
- You want a model that is still interpretable

### Applications

- Forecasting demand/sales trends
- Environmental modeling (e.g., rainfall vs yield)
- Medical progression patterns
- Engineering curves (e.g., stress-strain relationships)

## Slide 38: Limitations of Polynomial Regression

| Limitation | Explanation |
|:--|:--|
| Overfitting risk | High-degree models may memorize noise |
| Outlier sensitivity | Extreme points can shift the curve strongly |
| Extrapolation weakness | Predictions far outside training range can be unreliable |
| Limited flexibility for very complex patterns | Some problems are better handled by tree-based or neural models |

### Practical Tip

Use polynomial regression as a strong baseline, then compare with alternative models when data complexity increases.

## Slide 40: Support Vector Machine (SVM) and SVR

### What is SVR?

Support Vector Regression (SVR) is the regression version of Support Vector Machines.
It predicts a **continuous target** by fitting a function that stays as flat as possible while allowing a tolerance margin.

### Where It Is Used

- Disease progression prediction
- Engineering curves (e.g., stress-strain)
- Demand and trend forecasting

### Key Idea

Instead of minimizing all errors directly, SVR uses an **epsilon-insensitive zone** where small errors are ignored.

## Slide 41: SVR Formulation and Equations

### Regression Function

$$
f(x)=w^T\phi(x)+b
$$

- $\phi(x)$ maps input to (possibly) higher-dimensional space.
- With kernels, we avoid computing $\phi(x)$ explicitly.

### Optimization Objective (Soft-Margin SVR)

$$
\min_{w,b,\xi_i,\xi_i^*} \; \frac{1}{2}\|w\|^2 + C\sum_{i=1}^{n}(\xi_i+\xi_i^*)
$$

Subject to:

$$
y_i - (w^T\phi(x_i)+b) \le \varepsilon + \xi_i
$$

$$
(w^T\phi(x_i)+b) - y_i \le \varepsilon + \xi_i^*
$$

$$
\xi_i,\xi_i^* \ge 0
$$

### Hyperparameters

| Hyperparameter | Role |
|:--|:--|
| $\varepsilon$ | Width of the no-penalty tube around prediction |
| $C$ | Penalty strength for points outside the tube |
| Kernel | Controls curve shape (linear, RBF, polynomial) |

## Slide 42: Why SVR?

| Aspect | Traditional Regression | SVR |
|:--|:--|:--|
| Error handling | Minimizes all residuals | Ignores errors inside $\varepsilon$-tube |
| Outlier sensitivity | Can be sensitive | More robust when tube and $C$ are tuned |
| Non-linear handling | Needs explicit feature engineering | Uses kernels directly |
| Large datasets | Usually faster for very large data | Can become expensive for very large sample sizes |

### Practical Notes

- Use **feature scaling** before SVR.
- Start with RBF kernel, then tune $C$, $\varepsilon$, and kernel parameters.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

np.random.seed(21)

# Synthetic non-linear regression data
X = np.linspace(-3, 3, 90).reshape(-1, 1)
y_true = 0.7 * X[:, 0]**3 - 0.8 * X[:, 0]
y = y_true + np.random.normal(0, 1.2, size=X.shape[0])

# Scale features for SVR
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_grid = np.linspace(X.min(), X.max(), 300).reshape(-1, 1)
X_grid_scaled = scaler.transform(X_grid)

# ---------------- Figure 1: Kernel comparison ----------------
models = {
    'Linear Kernel': SVR(kernel='linear', C=10, epsilon=0.5),
    'RBF Kernel': SVR(kernel='rbf', C=10, epsilon=0.5, gamma=0.8),
    'Polynomial Kernel (deg=3)': SVR(kernel='poly', degree=3, C=10, epsilon=0.5, gamma='scale')
}

plt.figure(figsize=(15, 4.5))
for i, (name, model) in enumerate(models.items(), start=1):
    model.fit(X_scaled, y)
    y_pred_grid = model.predict(X_grid_scaled)

    plt.subplot(1, 3, i)
    plt.scatter(X[:, 0], y, color='royalblue', s=22, alpha=0.8, label='Data')
    plt.plot(X_grid[:, 0], y_pred_grid, color='crimson', linewidth=2.2, label='SVR fit')
    plt.title(name)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(alpha=0.25)
    plt.legend()

plt.suptitle('SVR with Different Kernels', fontsize=14, y=1.03)
plt.tight_layout()
plt.show()

# ---------------- Figure 2: Effect of epsilon and C ----------------
model_a = SVR(kernel='rbf', C=1, epsilon=1.2, gamma=0.8)   # wider tube, softer penalty
model_b = SVR(kernel='rbf', C=50, epsilon=0.2, gamma=0.8)  # narrower tube, stronger penalty

model_a.fit(X_scaled, y)
model_b.fit(X_scaled, y)

y_a = model_a.predict(X_grid_scaled)
y_b = model_b.predict(X_grid_scaled)

plt.figure(figsize=(12, 5))

# Left: wide epsilon, smaller C
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], y, color='royalblue', s=22, alpha=0.8)
plt.plot(X_grid[:, 0], y_a, color='seagreen', linewidth=2.2, label='SVR fit')
plt.fill_between(X_grid[:, 0], y_a - model_a.epsilon, y_a + model_a.epsilon,
                 color='seagreen', alpha=0.15, label='epsilon-tube')
plt.title('Higher epsilon, Lower C')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.25)
plt.legend()

# Right: narrow epsilon, larger C
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], y, color='royalblue', s=22, alpha=0.8)
plt.plot(X_grid[:, 0], y_b, color='darkorange', linewidth=2.2, label='SVR fit')
plt.fill_between(X_grid[:, 0], y_b - model_b.epsilon, y_b + model_b.epsilon,
                 color='darkorange', alpha=0.15, label='epsilon-tube')
plt.title('Lower epsilon, Higher C')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.25)
plt.legend()

plt.suptitle('SVR Hyperparameters: epsilon and C', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

## Slide 43: Decision Tree Regression (CART)

### What is Decision Tree Regressor?

Decision Tree Regressor is a supervised learning model used to predict **continuous values**.

Instead of fitting one global equation, it splits the feature space into regions using rule-based decisions, then predicts a constant value in each region.

### Why It Is Useful

- Captures non-linear relationships
- Easy to interpret (if tree depth is controlled)
- Handles complex interactions between features

## Slide 44: Core Concept of Regression Trees

A regression tree builds a flowchart-like structure:

- **Internal node:** decision rule (e.g., `x <= 4.5`)
- **Branch:** outcome of the rule
- **Leaf node:** predicted value (usually the mean target of that region)

### CART Perspective

| Tree Type | Target Type | Typical Criterion |
|:--|:--|:--|
| Classification Tree | Categorical target | Gini / Entropy |
| Regression Tree | Continuous target | MSE (or MAE) |

So in this section, we focus on the **Regression Tree** part of CART.

## Slide 45: How Splitting Happens (MSE Criterion)

At each node, the model tests candidate splits and selects the one with the **lowest weighted MSE**.

### Node MSE

$$
\text{MSE}(S)=\frac{1}{|S|}\sum_{i \in S}(y_i-\bar{y}_S)^2
$$

### Split Score

For split into left/right subsets $S_L, S_R$:

$$
\text{MSE}_{\text{split}}=\frac{|S_L|}{|S|}\text{MSE}(S_L)+\frac{|S_R|}{|S|}\text{MSE}(S_R)
$$

Choose the split with the minimum $\text{MSE}_{\text{split}}$.

### Prediction at Leaf

$$
\hat{y}_{\text{leaf}} = \frac{1}{|S_{leaf}|}\sum_{i\in S_{leaf}} y_i
$$

## Slide 46: Decision Tree Regression Algorithm (Simple Steps)

1. Start with all data at the root node.
2. Evaluate candidate splits across features.
3. Compute weighted MSE for each split.
4. Select the split with minimum error.
5. Repeat recursively on child nodes.
6. Stop when stopping rules are met (e.g., max depth, min samples).
7. Predict using the mean value at the reached leaf.

### Important Hyperparameters

| Hyperparameter | Effect |
|:--|:--|
| `max_depth` | Controls model complexity |
| `min_samples_split` | Minimum samples required to split a node |
| `min_samples_leaf` | Minimum samples in each leaf |
| `max_leaf_nodes` | Limits number of terminal regions |

These settings help reduce overfitting and improve generalization.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree

np.random.seed(10)

# Synthetic 1D non-linear data
X = np.sort(10 * np.random.rand(120, 1), axis=0)
y = np.sin(X).ravel() + 0.15 * np.random.randn(120)

x_grid = np.linspace(0, 10, 500).reshape(-1, 1)

# Trees with different complexity
tree_shallow = DecisionTreeRegressor(max_depth=2, random_state=10)
tree_medium = DecisionTreeRegressor(max_depth=4, random_state=10)
tree_deep = DecisionTreeRegressor(max_depth=10, random_state=10)

tree_shallow.fit(X, y)
tree_medium.fit(X, y)
tree_deep.fit(X, y)

y_shallow = tree_shallow.predict(x_grid)
y_medium = tree_medium.predict(x_grid)
y_deep = tree_deep.predict(x_grid)

# ---------------- Figure 1: Piecewise prediction (before/after complexity) ----------------
plt.figure(figsize=(13, 4.5))

plt.subplot(1, 3, 1)
plt.scatter(X, y, s=16, color='royalblue', alpha=0.7)
plt.plot(x_grid, y_shallow, color='crimson', linewidth=2.2)
plt.title('Shallow Tree (max_depth=2)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.25)

plt.subplot(1, 3, 2)
plt.scatter(X, y, s=16, color='royalblue', alpha=0.7)
plt.plot(x_grid, y_medium, color='seagreen', linewidth=2.2)
plt.title('Balanced Tree (max_depth=4)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.25)

plt.subplot(1, 3, 3)
plt.scatter(X, y, s=16, color='royalblue', alpha=0.7)
plt.plot(x_grid, y_deep, color='darkorange', linewidth=2.2)
plt.title('Deep Tree (max_depth=10)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.25)

plt.suptitle('Decision Tree Regression: Underfit -> Good Fit -> Overfit', fontsize=14, y=1.03)
plt.tight_layout()
plt.show()

# ---------------- Figure 2: Tree structure visualization ----------------
plt.figure(figsize=(14, 7))
plot_tree(
    tree_medium,
    filled=True,
    rounded=True,
    feature_names=['x'],
    fontsize=9
)
plt.title('Regression Tree Structure (CART)')
plt.show()

# ---------------- Figure 3: MSE split intuition (manual split demo) ----------------
x = X[:, 0]
threshold = 5.0
left_mask = x <= threshold
right_mask = x > threshold

left_mse = np.mean((y[left_mask] - np.mean(y[left_mask]))**2)
right_mse = np.mean((y[right_mask] - np.mean(y[right_mask]))**2)
weighted_mse = (left_mask.mean() * left_mse) + (right_mask.mean() * right_mse)

plt.figure(figsize=(10, 4.5))
plt.scatter(x, y, s=16, color='royalblue', alpha=0.75, label='Data')
plt.axvline(threshold, color='crimson', linestyle='--', linewidth=2, label=f'Split: x <= {threshold}')
plt.hlines(np.mean(y[left_mask]), xmin=x.min(), xmax=threshold, colors='seagreen', linewidth=2.2, label='Left leaf mean')
plt.hlines(np.mean(y[right_mask]), xmin=threshold, xmax=x.max(), colors='darkorange', linewidth=2.2, label='Right leaf mean')

plt.text(
    0.02, 0.97,
    f'Left MSE = {left_mse:.4f}\nRight MSE = {right_mse:.4f}\nWeighted Split MSE = {weighted_mse:.4f}',
    transform=plt.gca().transAxes,
    va='top',
    bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.9)
)

plt.title('Split Criterion Intuition (MSE Reduction)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.25)
plt.legend(loc='lower left')
plt.tight_layout()
plt.show()
```

## Slide 47: Random Forest Regression

### What is Random Forest Regressor?

Random Forest Regression is an ensemble method that combines many decision trees and averages their predictions.

### Algorithm (Simplified)

1. Randomly sample training data (bootstrap sample).
2. Build a decision tree on that sample.
3. Repeat to build $N_{trees}$ trees.
4. For a new input, let each tree predict $\hat{y}$.
5. Final prediction = average of all tree predictions.

$$
\hat{y}_{RF}(x)=\frac{1}{N_{trees}}\sum_{t=1}^{N_{trees}}\hat{y}_t(x)
$$

This averaging reduces variance and usually improves generalization.

## Slide 48: Decision Tree vs Random Forest (Regression)

| Aspect | Decision Tree Regressor | Random Forest Regressor |
|:--|:--|:--|
| Model type | Single tree | Ensemble of many trees |
| Split criterion | MSE minimization per split | Trees split by MSE, then averaged |
| Overfitting risk | Higher | Lower (variance reduction) |
| Stability | Sensitive to small data changes | More stable |
| Final prediction | Output of one tree | Mean of all tree outputs |
| Interpretability | High (simple tree) | Medium (many trees) |

### Practical Rule

- Use a single tree for interpretability.
- Use random forest when you need stronger predictive performance and robustness.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

np.random.seed(12)

# Synthetic non-linear regression data
X = np.linspace(0, 12, 220).reshape(-1, 1)
y = np.sin(X[:, 0]) + 0.2 * X[:, 0] + np.random.normal(0, 0.25, size=X.shape[0])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12)

# Models
dt = DecisionTreeRegressor(max_depth=5, random_state=12)
rf = RandomForestRegressor(n_estimators=250, max_depth=5, random_state=12)

dt.fit(X_train, y_train)
rf.fit(X_train, y_train)

# Predictions on smooth grid
x_grid = np.linspace(X.min(), X.max(), 500).reshape(-1, 1)
y_dt = dt.predict(x_grid)
y_rf = rf.predict(x_grid)

# Test metrics
dt_rmse = np.sqrt(mean_squared_error(y_test, dt.predict(X_test)))
rf_rmse = np.sqrt(mean_squared_error(y_test, rf.predict(X_test)))
dt_r2 = r2_score(y_test, dt.predict(X_test))
rf_r2 = r2_score(y_test, rf.predict(X_test))

# ---------- Plot 1: model curves ----------
plt.figure(figsize=(12, 4.8))
plt.scatter(X_train[:, 0], y_train, s=16, color='royalblue', alpha=0.45, label='Train Data')
plt.scatter(X_test[:, 0], y_test, s=20, color='black', alpha=0.6, label='Test Data')
plt.plot(x_grid[:, 0], y_dt, color='crimson', linewidth=2.2, label='Decision Tree Prediction')
plt.plot(x_grid[:, 0], y_rf, color='seagreen', linewidth=2.4, label='Random Forest Prediction')
plt.title('Decision Tree vs Random Forest (Regression)')
plt.xlabel('Feature x')
plt.ylabel('Target y')
plt.grid(alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()

# ---------- Plot 2: metric comparison ----------
labels = ['RMSE (lower better)', 'R² (higher better)']
dt_vals = [dt_rmse, dt_r2]
rf_vals = [rf_rmse, rf_r2]

x_pos = np.arange(len(labels))
width = 0.34

plt.figure(figsize=(8.5, 4.8))
plt.bar(x_pos - width/2, dt_vals, width=width, color='crimson', alpha=0.85, label='Decision Tree')
plt.bar(x_pos + width/2, rf_vals, width=width, color='seagreen', alpha=0.85, label='Random Forest')

for i, v in enumerate(dt_vals):
    plt.text(i - width/2, v + (0.02 if i == 1 else 0.01), f'{v:.3f}', ha='center', fontsize=9)
for i, v in enumerate(rf_vals):
    plt.text(i + width/2, v + (0.02 if i == 1 else 0.01), f'{v:.3f}', ha='center', fontsize=9)

plt.xticks(x_pos, labels)
plt.title('Model Evaluation on Test Set')
plt.grid(axis='y', alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()
```

## Slide 49: Evaluating Regression Models

### Key Performance Metrics

| Metric | Formula | What It Tells Us |
|:--|:--|:--|
| MAE | $\frac{1}{n}\sum \lvert y_i-\hat{y}_i \rvert$ | Average prediction error magnitude |
| MSE | $\frac{1}{n}\sum (y_i-\hat{y}_i)^2$ | Gives larger penalties to big errors |
| RMSE | $\sqrt{\frac{1}{n}\sum (y_i-\hat{y}_i)^2}$ | Error size in the original target scale |
| $R^2$ | $1-\frac{SS_{res}}{SS_{tot}}$ | Fraction of variance explained by the model |

### Best Practice

Evaluate models using more than one metric (for example, RMSE with $R^2$) to get a balanced view of performance.

## Slide 50: Regularization Methods (Why Needed?)

### Overfitting Problem

When a model is too flexible, it may fit noise instead of true patterns.

**Effects:**
- Very low training error
- Weak test performance
- High variance and unstable predictions

### General Regularized Objective

$$
\min_{\beta} \; \underbrace{\sum_{i=1}^{n}(y_i-\hat{y}_i)^2}_{\text{data loss}} + \lambda \cdot \underbrace{\Omega(\beta)}_{\text{penalty}}
$$

Regularization controls model complexity by penalizing large coefficients.

## Slide 51: Ridge, Lasso, and Elastic Net

### Without Regularization

$$
\min_{\beta} \sum_{i=1}^{n}(y_i-\hat{y}_i)^2
$$

### Ridge Regression (L2)

$$
\min_{\beta} \sum_{i=1}^{n}(y_i-\hat{y}_i)^2 + \lambda\sum_{j=1}^{m}\beta_j^2
$$

### Lasso Regression (L1)

$$
\min_{\beta} \sum_{i=1}^{n}(y_i-\hat{y}_i)^2 + \lambda\sum_{j=1}^{m}|\beta_j|
$$

### Elastic Net (L1 + L2)

$$
\min_{\beta} \sum_{i=1}^{n}(y_i-\hat{y}_i)^2 + \lambda_1\sum_{j=1}^{m}|\beta_j| + \lambda_2\sum_{j=1}^{m}\beta_j^2
$$

## Slide 52: Regularization Comparison Table

| Method | Penalty Type | Main Behavior | Feature Selection |
|:--|:--|:--|:--:|
| Ridge | L2 ($\sum \beta_j^2$) | Shrinks coefficients smoothly | No |
| Lasso | L1 ($\sum |\beta_j|$) | Shrinks and can set some coefficients to zero | Yes |
| Elastic Net | L1 + L2 | Combines shrinkage + selection | Yes |

### When to Use

- **Ridge:** many small/medium useful features.
- **Lasso:** only few important features expected.
- **Elastic Net:** correlated features and need both stability and sparsity.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

np.random.seed(5)

# Synthetic curved data (to show overfitting + regularization)
X = np.linspace(-3, 3, 220).reshape(-1, 1)
y_clean = 0.8 * X[:, 0]**3 - 1.5 * X[:, 0]
y = y_clean + np.random.normal(0, 2.0, size=X.shape[0])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=5)
X_plot = np.linspace(-3, 3, 500).reshape(-1, 1)

# High-degree polynomial to expose overfitting
poly_deg = 12

models = {
    'No Reg': make_pipeline(PolynomialFeatures(poly_deg, include_bias=False), StandardScaler(), LinearRegression()),
    'Ridge': make_pipeline(PolynomialFeatures(poly_deg, include_bias=False), StandardScaler(), Ridge(alpha=1.5)),
    'Lasso': make_pipeline(PolynomialFeatures(poly_deg, include_bias=False), StandardScaler(), Lasso(alpha=0.02, max_iter=50000)),
    'ElasticNet': make_pipeline(PolynomialFeatures(poly_deg, include_bias=False), StandardScaler(), ElasticNet(alpha=0.02, l1_ratio=0.6, max_iter=50000))
}

pred_curves = {}
train_rmse = {}
test_rmse = {}
nonzero_coef = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    pred_curves[name] = model.predict(X_plot)

    y_tr_pred = model.predict(X_train)
    y_te_pred = model.predict(X_test)
    train_rmse[name] = np.sqrt(mean_squared_error(y_train, y_tr_pred))
    test_rmse[name] = np.sqrt(mean_squared_error(y_test, y_te_pred))

    # Count non-zero coefficients (for feature-selection intuition)
    reg = model.named_steps[list(model.named_steps.keys())[-1]]
    if hasattr(reg, 'coef_'):
        nonzero_coef[name] = int(np.sum(np.abs(reg.coef_) > 1e-6))
    else:
        nonzero_coef[name] = np.nan

# -------- Figure 1: prediction curves --------
plt.figure(figsize=(13, 8))

plot_order = ['No Reg', 'Ridge', 'Lasso', 'ElasticNet']
colors = {'No Reg': 'crimson', 'Ridge': 'seagreen', 'Lasso': 'darkorange', 'ElasticNet': 'purple'}

for i, name in enumerate(plot_order, start=1):
    plt.subplot(2, 2, i)
    plt.scatter(X_train[:, 0], y_train, s=14, color='royalblue', alpha=0.45, label='Train')
    plt.scatter(X_test[:, 0], y_test, s=18, color='black', alpha=0.55, label='Test')
    plt.plot(X_plot[:, 0], pred_curves[name], color=colors[name], linewidth=2.2, label=name)
    plt.plot(X_plot[:, 0], 0.8 * X_plot[:, 0]**3 - 1.5 * X_plot[:, 0], color='gray', linestyle='--', linewidth=1.4, label='True trend')
    plt.title(f'{name} (degree={poly_deg})')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(alpha=0.25)
    plt.legend(fontsize=8)

plt.suptitle('Regularization Effect on Polynomial Regression', fontsize=14, y=1.01)
plt.tight_layout()
plt.show()

# -------- Figure 2: train/test RMSE comparison --------
labels = plot_order
x_idx = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(9.5, 4.8))
plt.bar(x_idx - width/2, [train_rmse[k] for k in labels], width=width, color='royalblue', alpha=0.85, label='Train RMSE')
plt.bar(x_idx + width/2, [test_rmse[k] for k in labels], width=width, color='crimson', alpha=0.85, label='Test RMSE')
plt.xticks(x_idx, labels)
plt.ylabel('RMSE')
plt.title('Overfitting Check: Train vs Test Error')
plt.grid(axis='y', alpha=0.25)
plt.legend()
plt.tight_layout()
plt.show()

# -------- Figure 3: non-zero coefficients --------
plt.figure(figsize=(8.5, 4.5))
vals = [nonzero_coef[k] for k in labels]
plt.bar(labels, vals, color=['crimson', 'seagreen', 'darkorange', 'purple'], alpha=0.9)
for i, v in enumerate(vals):
    plt.text(i, v + 0.2, str(v), ha='center', fontsize=9)
plt.title('Model Sparsity (Number of Non-Zero Coefficients)')
plt.ylabel('Count of Non-Zero Coefficients')
plt.grid(axis='y', alpha=0.25)
plt.tight_layout()
plt.show()
```

## Slide 53: Regression Models - Advantages and Disadvantages

| Model | Advantages | Disadvantages |
|:--|:--|:--|
| Linear Regression | Simple, fast, easy to interpret, works well for linear patterns. | Assumes linear relationship and is sensitive to outliers. |
| Polynomial Regression | Captures curved/non-linear relationships better than linear regression. | Choosing degree is sensitive; high degrees may overfit. |
| SVR | Handles non-linear patterns with kernels and can be robust to noise with proper tuning. | Requires feature scaling and careful hyperparameter tuning; slower on large datasets. |
| Decision Tree Regression | Interpretable rules, captures non-linear interactions, no feature scaling required. | Can overfit easily and may be unstable with small data changes. |
| Random Forest Regression | Higher accuracy and stability, reduces overfitting by averaging many trees. | Less interpretable and can be computationally heavier. |

### Quick Selection Tip

- If interpretability is a priority: **Linear Regression** or **Decision Tree**.
- If predictive performance is a priority: **Random Forest**.
- If the relationship is strongly non-linear with careful tuning: **SVR** or **Polynomial Regression**.

### Model Summary (Name + Equation)

| Model | Core Equation |
|:--|:--|
| Linear Regression | $\hat{y} = b_0 + b_1x$ |
| Polynomial Regression | $\hat{y} = b_0 + b_1x + b_2x^2 + \cdots + b_dx^d$ |
| SVR | $f(x)=w^T\phi(x)+b$ |
| Decision Tree Regression | $\hat{y}_{leaf}=\frac{1}{\lvert S_{leaf} \rvert}\sum_{i\in S_{leaf}} y_i$ |
| Random Forest Regression | $\hat{y}_{RF}(x)=\frac{1}{N_{trees}}\sum_{t=1}^{N_{trees}}\hat{y}_t(x)$ |

## Slide 54: Visual Comparison of Regression Models

The following figure shows one plot for each model using the same dataset:

- Linear Regression
- Polynomial Regression
- SVR
- Decision Tree Regression
- Random Forest Regression

This helps compare how each model fits linear vs non-linear patterns.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

np.random.seed(14)

# Same dataset for all models (non-linear pattern)
X = np.linspace(-3, 3, 120).reshape(-1, 1)
y = 0.8 * X[:, 0]**3 - 1.4 * X[:, 0] + np.random.normal(0, 1.8, size=X.shape[0])

X_grid = np.linspace(X.min(), X.max(), 400).reshape(-1, 1)

models = {
    'Linear Regression': LinearRegression(),
    'Polynomial Regression (deg=3)': make_pipeline(PolynomialFeatures(degree=3, include_bias=False), LinearRegression()),
    'SVR (RBF)': make_pipeline(StandardScaler(), SVR(kernel='rbf', C=12, epsilon=0.4, gamma='scale')),
    'Decision Tree Regression': DecisionTreeRegressor(max_depth=4, random_state=14),
    'Random Forest Regression': RandomForestRegressor(n_estimators=220, max_depth=4, random_state=14)
}

preds = {}
eq_map = {}


def fmt_signed(v):
    return f"{v:+.2f}"


for name, model in models.items():
    model.fit(X, y)
    preds[name] = model.predict(X_grid)

    # Build equation text for each model
    if name == 'Linear Regression':
        reg = model
        b0 = reg.intercept_
        b1 = reg.coef_[0]
        eq_map[name] = rf"$\hat{{y}}={b0:.2f}{fmt_signed(b1)}x$"

    elif name == 'Polynomial Regression (deg=3)':
        reg = model.named_steps['linearregression']
        # Coefficients correspond to x, x^2, x^3
        b0 = reg.intercept_
        b1, b2, b3 = reg.coef_[:3]
        eq_map[name] = (
            rf"$\hat{{y}}={b0:.2f}{fmt_signed(b1)}x{fmt_signed(b2)}x^2{fmt_signed(b3)}x^3$"
        )

    elif name == 'SVR (RBF)':
        eq_map[name] = r"$f(x)=\sum_i(\alpha_i-\alpha_i^*)K(x_i,x)+b$"

    elif name == 'Decision Tree Regression':
        eq_map[name] = r"$\hat{y}_{leaf}=\frac{1}{|S_{leaf}|}\sum_{i\in S_{leaf}} y_i$"

    elif name == 'Random Forest Regression':
        eq_map[name] = r"$\hat{y}_{RF}(x)=\frac{1}{N_{trees}}\sum_{t=1}^{N_{trees}}\hat{y}_t(x)$"

fig, axes = plt.subplots(2, 3, figsize=(16, 10.5), constrained_layout=True)
axes = axes.ravel()

colors = {
    'Linear Regression': 'crimson',
    'Polynomial Regression (deg=3)': 'seagreen',
    'SVR (RBF)': 'darkorange',
    'Decision Tree Regression': 'purple',
    'Random Forest Regression': 'teal'
}

for i, (name, y_pred) in enumerate(preds.items()):
    ax = axes[i]
    ax.scatter(X[:, 0], y, color='royalblue', s=18, alpha=0.65, label='Data')
    ax.plot(X_grid[:, 0], y_pred, color=colors[name], linewidth=2.4, label=name)

    # Special visualization for SVR: epsilon tube + support vectors
    if name == 'SVR (RBF)':
        svr_pipe = models[name]
        svr_model = svr_pipe.named_steps['svr']
        scaler_model = svr_pipe.named_steps['standardscaler']

        # Epsilon-insensitive tube around prediction
        eps = svr_model.epsilon
        ax.fill_between(
            X_grid[:, 0],
            y_pred - eps,
            y_pred + eps,
            color='darkorange',
            alpha=0.28,
            label='epsilon-tube'
        )
        ax.plot(X_grid[:, 0], y_pred + eps, color='darkorange', linestyle='--', linewidth=1.3)
        ax.plot(X_grid[:, 0], y_pred - eps, color='darkorange', linestyle='--', linewidth=1.3)

        # Support vectors (transform back to original x scale)
        sv_x = scaler_model.inverse_transform(svr_model.support_vectors_)[:, 0]
        sv_y = y[svr_model.support_]
        ax.scatter(
            sv_x,
            sv_y,
            facecolors='none',
            edgecolors='black',
            s=70,
            linewidths=1.2,
            label='Support vectors'
        )

    ax.set_title(name)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.grid(alpha=0.25)
    ax.legend(fontsize=8)

    # Equation text under each subplot
    ax.text(
        0.5,
        0.06,
        eq_map[name],
        transform=ax.transAxes,
        ha='center',
        va='bottom',
        fontsize=9,
        bbox=dict(boxstyle='round,pad=0.25', facecolor='white', alpha=0.92, edgecolor='gray')
    )

# Hide the 6th empty subplot
axes[-1].axis('off')

fig.suptitle('One Plot per Regression Model', fontsize=16)
plt.show()
```
